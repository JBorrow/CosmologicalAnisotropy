```python
%pylab inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
from pyanis import arrayofanis
from reader import *

radra+=pi
```

```python
# Regular lambda-CDM

vd = vectorize(dl)
lumdists = vd(zs, omegam)
dmu = mus - muth(lumdists)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(radra, raddec, dmu)
```

```python
output = arrayofanis(radra, raddec, dmu)
```

```python
figure()
imshow(output.T, interpolation='none', cmap='viridis')
colorbar()
```

```python
# Linear correlation
figure()
inputra = arange(0, 2*pi, 0.01)
inputdec = arange(-0.5*pi, 0.5*pi, 0.005)
dmu = inputra
output = arrayofanis(inputra, inputdec, dmu)

imshow(output.T, cmap='viridis')
colorbar()
```

```python
# Now try with dipole prediction and real data

from dipoler import muthdip

newmuths = muthdip(muth(lumdists), radra, raddec, 0.1, 0.1, 0.001)
fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(radra, raddec, mus-newmuths)
```

```python
figure()
dmu = mus - newmuths
output = arrayofanis(radra, raddec, dmu)

imshow(output.T, cmap='viridis')
colorbar()
```

```python
scatter(raddec, newmuths)
scatter(raddec, muth(lumdists), color='r')
```

```python
#now what about purely theoretical data?

figure()
dmu = muth(lumdists) - newmuths
output = arrayofanis(radra, raddec, dmu)

imshow(output.T, cmap='viridis')
colorbar()
```

```python
#now what about purely theoretical data?

fig = figure(figsize=(16, 8))
ax = subplot(111, projection='mollweide')
dmu = muth(lumdists) - newmuths
output = arrayofanis(radra, raddec, dmu)

img = ax.imshow(output.T, cmap='viridis', extent=[-pi, pi, -pi/2, pi/2])
colorbar(img)
```
