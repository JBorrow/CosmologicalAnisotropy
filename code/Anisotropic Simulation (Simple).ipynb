```python
%pylab inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
from reader import *

omegam = 0.288
```

```python
ra = numpy.random.rand(740)*2*pi - pi
dec = numpy.random.rand(740)*pi - pi/2
```

Let's have a dipole!

```python
nhat = array([sin(dec+pi/2)*cos(ra+pi),sin(dec+pi/2)*sin(ra+pi),cos(dec+pi/2)])
dipstr = dot(array([1, 0, 0]), nhat)
```

```python
dmu = dipstr + numpy.random.rand(740)
```

```python
from pyanis import arrayofanis
```

```python
figure()
ax = subplot(projection='mollweide')
scatter(ra,dec,c=dmu)
```

```python
output = arrayofanis(ra+pi, dec, dmu)
```

```python
figure()
imshow(output.T, interpolation='none', cmap='viridis')
colorbar()
```
